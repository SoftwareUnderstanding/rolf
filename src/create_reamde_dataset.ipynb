```python
import csv
import pandas as pd
from Preprocessor1 import Preprocessor
```

```python
reader = csv.DictReader(open('../documentation/dataset/awesome_links.csv'), delimiter=';')
#data = {key: [] for key in reader.fieldnames}
data = {key: [] for key in ['Label', 'Repo', 'Text']}
for row in reader:
	filename = '_'.join(row['Repo'].split('/')[-2:]) + '.txt'
	print(filename)
	with open('../data/awesome_lists_links/readme/semantic_web/' + filename) as f:
		#print(f.readlines())
		text = ' '.join(f.readlines())
		text = text.replace('\n', ' ')
		data['Text'].append('"' + text + '"')
	data['Repo'].append(row['Repo'])
	data['Label'].append(row['Label'])

df = pd.DataFrame(data)
#Preprocessor(df).run()
df.drop_duplicates(inplace=True)
df.to_csv("../data/categories/readme_semantic_web.csv", sep=';', index=False)
```
