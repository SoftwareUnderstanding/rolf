```python
import csv
import pandas as pd
from Preprocessor import Preprocessor
```

```python
reader = csv.DictReader(open('../data/awesome_lists_links/repos.csv'), delimiter=';')
#data = {key: [] for key in reader.fieldnames}
data = {key: [] for key in ['Label', 'Repo', 'Text']}
for row in reader:
	filename = '_'.join(row['Repo'].split('/')[-2:]) + '.txt'
	print(filename)
	with open('../data/somef_dataset/papers_with_code/' + filename) as f:
		#print(f.readlines())
		text = ' '.join(f.readlines())
		text = text.replace('\n', ' ')
		data['Text'].append('"' + text + '"')
	data['Repo'].append(row['Repo'])
	data['Label'].append(row['Label'])

df = pd.DataFrame(data)
#Preprocessor(df).run()
df.drop_duplicates(inplace=True)
df.to_csv("../data/readme.csv", sep=';', index=False)
```
