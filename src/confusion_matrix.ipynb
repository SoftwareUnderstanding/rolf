```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../results/results/readme_semantic_web_train_readme_semantic_web_test.csv', sep=';')
df.set_index(['PipelineID'], inplace=True)
```

```python
df_cm = df[['test_tp_mean', 'test_fn_mean', 'test_fp_mean', 'test_tn_mean']]
```

```python
df_cm.head(10)
```

```python
for ind, row in df_cm.iterrows():
    print(list(row))
    data = np.asarray(list(row)).reshape(2,2)
    plt.figure(figsize=(5,4))
    sns.heatmap(data, annot=True)
    plt.title('Confusion Matrix '+ind)
    plt.ylabel('Actual Values')
    plt.xlabel('Predicted Values')
    plt.show()
```
