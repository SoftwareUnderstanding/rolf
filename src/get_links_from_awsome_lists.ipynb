```python
import re
import pandas as pd
```

```python
regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
```

```python
lines = []
with open("awesome_lists/semantic_web.txt", "r") as f:
    lines = f.read().split('\n')
```

```python
github_links = []
for line in lines:
    if line != '':
        urls = re.findall(regex, line)
        for url in urls:
            if 'github' in url[0]: 
                l = url[0].split('/')
                if len(l) >= 5:
                    link = l[0] + '/' + l[1] + '/' + l[2] + '/' + l[3] + '/' + l[4] 
                    if link not in github_links:
                        github_links.append(url[0])
```

```python
keys = ['Label', 'Repo']
data = { key: [] for key in keys}
for row in github_links:
    data['Label'].append('Semantic Web')
    data['Repo'].append(row)
```

```python
print(len(github_links))
df = pd.DataFrame(data)
df.to_csv('dataset/awesome_links.csv', index=False, sep=';')
```
