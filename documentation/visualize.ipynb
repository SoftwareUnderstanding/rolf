```python
for r in res:
    for key in r.keys():
        r[key]['f1'] = (2*r[key]["precision"]*r[key]["recall"])/(r[key]["precision"]+r[key]["recall"])
        print(r[key]['f1'])
print(res)
```

```python
precision = []
recall = []
f1 = []
threshold = []
for r in res:
    for key in r.keys():
        precision.append(r[key]["precision"])
        recall.append(r[key]["recall"])
        f1.append(r[key]["f1"])
        threshold.append(float(key))
```

```python
#plt.figure(figsize=(5,4))
plt.plot(threshold, precision, label="Precision", color="red")
plt.plot(threshold, recall, label="Recall", color="green")
plt.plot(threshold, f1, label="F1-score", color="blue")
plt.title("Results by threshold for the ROLF_BERT approach", size=12)
plt.xlabel("Threshold", size=12)
plt.legend(loc='best')
plt.xlim([-0.01, 1.01])
plt.ylim([0.5, 1.1])
plt.savefig("bert_results.png")
```
